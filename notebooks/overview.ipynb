```python
import matplotlib.pyplot as plt
import numpy as np

# Data
item_one = {'name': 'svm', 'precision': 0.98, 'recall': 0.98, 'f1': 0.50}
item_two = {'name': 'knn', 'precision': 0.98, 'recall': 0.98, 'f1': 0.60}

names = [item_one['name'], item_two['name']]
precision_values = [item_one['precision'], item_two['precision']]
recall_values = [item_one['recall'], item_two['recall']]
f1_values = [item_one['f1'], item_two['f1']]

num_items = len(names)
bar_width = 0.25
bar_offsets = np.arange(len(names)) * bar_width * 4

ind = np.arange(num_items)

# Plotting
plt.figure(figsize=(10, 8))

plt.bar(bar_offsets, precision_values, width=bar_width, label='Precision')
plt.bar(bar_offsets + bar_width, recall_values, width=bar_width, label='Recall')
plt.bar(bar_offsets + 2 * bar_width, f1_values, width=bar_width, label='F1')

plt.xlabel('Models')
plt.ylabel('Score')
plt.title('Metric Comparison for SVM and KNN')
plt.xticks(bar_offsets + bar_width, names)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Data
item_one = {'name': 'svm', 'precision': 0.98, 'recall': 0.95, 'f1': 0.96}
item_two = {'name': 'knn', 'precision': 0.98, 'recall': 0.96, 'f1': 0.96}

names = [item_one['name'], item_two['name']]
precision_values = [item_one['precision'], item_two['precision']]
recall_values = [item_one['recall'], item_two['recall']]
f1_values = [item_one['f1'], item_two['f1']]

num_items = len(names)
bar_width = 0.5  # Width of each bar
bar_offsets = np.arange(len(names)) * bar_width * 2  # Adjust offset to position bars for each metric

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Bar heights (metric values)
precision_heights = precision_values
recall_heights = recall_values
f1_heights = f1_values

# X positions for bars
x_pos_precision = bar_offsets
x_pos_recall = bar_offsets + bar_width
x_pos_f1 = bar_offsets + 2 * bar_width

# Y positions for bars (model names)
y_pos = np.arange(num_items)

# Plot bars
ax.bar3d(x_pos_precision, y_pos, np.zeros(num_items), bar_width, bar_width, precision_heights, label='Precision', color='b', zsort='average')
ax.bar3d(x_pos_recall, y_pos, np.zeros(num_items), bar_width, bar_width, recall_heights, label='Recall', color='g', zsort='average')
ax.bar3d(x_pos_f1, y_pos, np.zeros(num_items), bar_width, bar_width, f1_heights, label='F1', color='r', zsort='average')

# Set labels and title
ax.set_xlabel('Models')
ax.set_ylabel('Metrics')
ax.set_zlabel('Score')
ax.set_title('3D Grouped Bar Chart')

# Set ticks and legend
ax.set_xticks(bar_offsets + bar_width)
ax.set_xticklabels(names)
ax.legend()

plt.show()
```
